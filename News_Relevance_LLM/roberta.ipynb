```python

# Get predicted probabilities
logits = outputs.logits
# probabilities = torch.softmax(logits, dim=1)
probabilities = tf.nn.softmax(logits, axis=1)
```

```python

# Mapping of class indices to labels
class_labels = {
    0: 'AI',
    1: 'Quantum Computing',
    2: 'Green Computing',
    3: 'Robotics',
    4: 'Trust Technologies',
    5: 'Anti-disinformation technologies',
    6: 'Communications Technologies',
}
```

```python

# Determine predicted classes for each article
predicted_classes = [tf.argmax(prob).numpy() for prob in probabilities]
```

```python

# Map predicted class indices to labels
predicted_labels = [class_labels.get(class_idx, 'General') for class_idx in predicted_classes]
```
